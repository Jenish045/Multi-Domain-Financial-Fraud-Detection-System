```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Load data
df = pd.read_csv('../data/raw/creditcard.csv')
print(f'Shape of dataset: {df.shape}')
```

```python
# Class distribution
dist = df['Class'].value_counts(normalize=True) * 100
print(dist)
sns.countplot(x='Class', data=df)
plt.title('Class Distribution (0: Normal, 1: Fraud)')
plt.show()
```

```python
# Amount distribution
plt.figure(figsize=(10, 6))
sns.histplot(df[df['Class'] == 0]['Amount'], bins=50, color='blue', log_scale=True, label='Normal', kde=True, alpha=0.5)
sns.histplot(df[df['Class'] == 1]['Amount'], bins=50, color='red', log_scale=True, label='Fraud', kde=True, alpha=0.5)
plt.title('Amount Distribution (Log Scale)')
plt.legend()
plt.show()
```

```python
# Correlation Heatmap
corr = df.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(corr, cmap='coolwarm', cbar=True)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Time-based pattern
df['hour'] = (df['Time'] % 86400) // 3600
fraud_rate = df.groupby('hour')['Class'].mean() * 100
plt.figure(figsize=(10, 5))
fraud_rate.plot(kind='bar', color='orange')
plt.title('Fraud Rate by Hour of Day')
plt.ylabel('Fraud Rate (%)')
plt.show()
```

```python
# Box plots for top correlated features
top_features = corr['Class'].abs().sort_values(ascending=False).index[1:6]
plt.figure(figsize=(15, 5))
for i, col in enumerate(top_features):
    plt.subplot(1, 5, i+1)
    sns.boxplot(x='Class', y=col, data=df)
    plt.title(f'{col} vs Class')
plt.tight_layout()
plt.show()
```
