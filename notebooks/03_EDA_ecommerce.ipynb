```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load ecommerce data
df = pd.read_csv('../data/raw/ecommerce_fraud.csv').sample(n=10000, random_state=42) # sample for fast EDA
print(f'Sample Shape: {df.shape}')
```

```python
# Fraud by Payment Method
payment_fraud = df.groupby('PaymentMethod')['Is_Fraudulent'].mean() * 100
payment_fraud.sort_values().plot(kind='barh', color='purple')
plt.title('Fraud Rate by Payment Method (%)')
plt.show()
```

```python
# Fraud by Device
sns.barplot(x='DeviceUsed', y='Is_Fraudulent', data=df)
plt.title('Fraud Rate by Device')
plt.ylabel('Fraud Rate')
plt.show()
```

```python
# Transaction amount
plt.figure(figsize=(8, 5))
sns.kdeplot(df[df['Is_Fraudulent']==0]['TransactionAmount'], label='Normal', shade=True)
sns.kdeplot(df[df['Is_Fraudulent']==1]['TransactionAmount'], label='Fraud', shade=True)
plt.title('Transaction Amount Distribution')
plt.legend()
plt.show()
```

```python
# Fraud by hour
df['TransactionHour'] = pd.to_datetime(df['Transaction Date']).dt.hour
hour_fraud = df.groupby('TransactionHour')['Is_Fraudulent'].mean()
hour_fraud.plot(kind='line', marker='o')
plt.title('Fraud Rate by Hour of Day')
plt.ylabel('Rate')
plt.show()
```

```python
# Account age boxplot
sns.boxplot(x='Is_Fraudulent', y='AccountAgeDays', data=df)
plt.title('Account Age vs Fraudulent Status')
plt.show()
```

```python
# Address mismatch
df['AddressMismatch'] = (df['shipping_address'] != df['billing_address']).astype(int)
sns.barplot(x='AddressMismatch', y='Is_Fraudulent', data=df)
plt.title('Fraud Rate vs Address Mismatch')
plt.show()
```
